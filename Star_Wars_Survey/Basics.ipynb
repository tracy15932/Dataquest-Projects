# Star War Survey

## Introduction:

This project is aiming to answer to question:
**does the rest of America realize that “The Empire Strikes Back” is clearly the best of the bunch?**

The dataset _star_wars_ includes 835 total survey responses from Star Wars fans.
- **RespondentID** - An anonymized ID for the respondent (person taking the survey)
- **Gender** - The respondent's gender
- **Age** - The respondent's age
- **Household Income** - The respondent's income
- **Education** - The respondent's education level
- **Location (Census Region)** - The respondent's location
- **Have you seen any of the 6 films in the Star Wars franchise?** - Has a Yes or No response
- **Do you consider yourself to be a fan of the Star Wars film franchise?** - Has a Yes or No respons
- **ranking1-6**- 1 means the film was the most favorite, and 6 means it was the least favorite

## Data Cleanning

```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")

star_wars.head(10)
```

```python
star_wars.columns
```

```python
## Remove rows with invalid RsspondentID
star_wars=star_wars[pd.notnull(star_wars['RespondentID'])]
star_wars.head()
```

```python
yes_no={
    "Yes":True,
    "No": False
}
cols=['Have you seen any of the 6 films in the Star Wars franchise?',
      'Do you consider yourself to be a fan of the Star Wars film franchise?'      
     ]
for col in cols:
    star_wars[col]=star_wars[col].map(yes_no)
star_wars.head()
```

```python
import numpy as np
cols=star_wars.columns[3:9]
movie_map={
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for each in cols:
    star_wars[each]=star_wars[each].map(movie_map)

star_wars=star_wars.rename(columns={
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    "Unnamed: 4":"seen_2",
    "Unnamed: 5":"seen_3",
    "Unnamed: 6":"seen_4",
    "Unnamed: 7":"seen_5",
    "Unnamed: 8":"seen_6"
})

star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
star_wars=star_wars.rename(columns={
    "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
    "Unnamed: 10": "ranking_2",
    "Unnamed: 11": "ranking_3",
    "Unnamed: 12": "ranking_4",
    "Unnamed: 13": "ranking_5",
    "Unnamed: 14": "ranking_6"
})

star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]].mean()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.bar(range(6), star_wars[star_wars.columns[9:15]].mean())
plt.show()
```

So far, we have cleaned the data, renamed some columns and compute the mean of rankings. From the plot above, we observe that the original movies are ranked higher than the newer ones.

```python
star_wars[star_wars.columns[3:8]].sum()
```

```python
plt.bar(range(5), star_wars[star_wars.columns[3:8]].sum())
plt.show()
```

More people watched the original movies than the newer ones. This reinforces our observation from rankings, which the earlier movies are more popular.

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

```python
plt.bar(range(6), males[males.columns[9:15]].mean())
plt.title('Ranking Male vs Female')
plt.show()
plt.bar(range(6), females[females.columns[9:15]].mean(),color='red')
plt.show()
```

```python
plt.bar(range(5), males[males.columns[3:8]].sum())
plt.title('Movies Seen Total- Male vs Female')
plt.show()
plt.bar(range(5), females[females.columns[3:8]].sum(),color='red')
plt.show()
```

More males watched the first three episodes than females. However, Females love the movies far more than males.
